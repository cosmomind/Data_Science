# Question 1: Use yfinance to Extract Stock Data

# Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
import yfinance as yf

# Create Tesla ticker object
tesla = yf.Ticker("TSLA")

# Get quarterly financials
financials = tesla.quarterly_financials

# Transpose to make dates rows
tesla_revenue = financials.loc["Total Revenue"].transpose()

# Reset index and rename columns
tesla_revenue = tesla_revenue.reset_index()
tesla_revenue.columns = ["Date", "Revenue"]

# Display last five rows
print(tesla_revenue.tail())

```

# Question 5: Plot Tesla Stock Graph



```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph():
    # Step 1: Create a Ticker object for Tesla
    tesla = yf.Ticker("TSLA")

    # Step 2: Get historical stock data
    tesla_data = tesla.history(period="max")

    # Step 3: Reset index so 'Date' becomes a column
    tesla_data.reset_index(inplace=True)

    # Step 4: Plot the closing price over time
    plt.figure(figsize=(12,6))
    plt.plot(tesla_data['Date'], tesla_data['Close'], label='Close Price', color='blue')
    plt.title('Tesla Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


make_graph()

```
